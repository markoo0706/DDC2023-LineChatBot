```python
def handle_df(df : dict):
    resName = [i for i in df.keys()]
    lat_list = [df[res]["lat"] for res in resName]
    lng_list = [df[res]["lng"] for res in resName]
    rating_list = [df[res]["rating"] for res in resName]
    label_list = [df[res]["label"] for res in resName]
    aspect_list = [df[res]["aspect"] for res in resName]
    colName = ["Name", "laitude", "longtitude", "rating", "label", "aspect"]
    data = {
    colName[0]: resName,
    colName[1]: lat_list,
    colName[2]: lng_list,
    colName[3]: rating_list,
    colName[4] : label_list,
    colName[5] : aspect_list
 }
    df_new = pd.DataFrame(data)
    return df_new
```

```python
def chang_df(df1, df2):
    resname = [i for i in df1.keys()]
    for res in resname:
        df1[res]["label"] = df2[res]["restaurant_label"]
        df1[res]["aspect"] = df2[res]["restaurant_aspect"] 
    return df1

df = chang_df(df1, df2)
df
```

```python
new_df = handle_df(df1)
new_df.to_csv("new_data.csv")
```
